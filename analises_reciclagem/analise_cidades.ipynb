```python
# importação das bibliotecas
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

# Preparação dos dados

```python
# carregamento dos dados
principais_cidades_Brasil = pd.read_csv('Principais_cidades_do_pais_em_reciclagem.csv')
```

```python
# tamanho do dataframe
principais_cidades_Brasil.shape
```

```python
# as primeiras linhas do dataframe
principais_cidades_Brasil.head()
```

```python
# informações gerais
principais_cidades_Brasil.info()
```

```python
# informações métricas
principais_cidades_Brasil.describe()
```

```python
# renomenando as colunas
principais_cidades_Brasil = principais_cidades_Brasil.rename(columns={"Cidade":"cidade", "Estado":"estado", "Materiais Reciclados":"mat_recl", 
                                                                         "Papel (%)": "papel", "Plástico (%)": "plastico", "Vidro (%)": "vidro", 
                                                                         "Alumínio (%)": "aluminio", "Metal (%)":"metal", })
```

```python
principais_cidades_Brasil.isna().sum()
```

```python
principais_cidades_Brasil.duplicated().sum()
```

**Overview:**
- Os nomes das colunas devem ficar em letra minúscula e devem ser claros;
- Não há dados duplicados ou dados ausentes;

# EDA

```python
principais_cidades_Brasil.sort_values('estado', inplace=True)
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))
sns.barplot(x='estado', y='papel', data=principais_cidades_Brasil, palette='crest')
plt.title('Principais Estados - papel')
plt.xticks(rotation=45)
plt.show()
```

Os estados que reciclam mais papel no Brasil são:
- MG
- SP
- PR
- SC
- ES
- GO

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))
sns.barplot(x='estado', y='plastico', data=principais_cidades_Brasil, palette='crest')
plt.title('Principais Estados - plástico')
plt.xticks(rotation=45)
plt.show()
```

```python
# gráfico de barras para o material metal
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))
sns.barplot(x='estado', y='metal', data=principais_cidades_Brasil, palette='crest')
plt.title('Principais Estados - metal')
plt.xticks(rotation=45)
plt.show()
```

Os estados reciclam metal igualmente no Brasil

# Conclusão:
