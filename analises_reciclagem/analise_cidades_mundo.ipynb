```python
# importação das bibliotecas
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# carregamento dos dados
principais_cidades_mundo = pd.read_csv('Principais_cidades_do_mundo_em_reciclagem.csv')
```

```python
# tamanho do dataframe
principais_cidades_mundo.shape
```

```python
# as primeiras linhas do dataframe
principais_cidades_mundo.head()
```

```python
# informações gerais
principais_cidades_mundo.info()
```

```python
# informações métricas
principais_cidades_mundo.describe()
```

```python
# renomenando as colunas
principais_cidades_mundo_df = principais_cidades_mundo.rename(columns={"Cidade":"cidade", "País":"pais", "Materiais Reciclados":"mat_recl", 
                                                                       "Papel (%)": "papel", "Plástico (%)": "plastico", "Vidro (%)": "vidro", 
                                                                       "Alumínio (%)": "aluminio", "Metal (%)":"metal"})
```

```python
principais_cidades_mundo_df.isna().sum()
```

```python
principais_cidades_mundo_df.duplicated().sum()
```

-----

```python
principais_cidades_mundo_df.sort_values('pais', inplace=True)
```

```python
principais_cidades_mundo_df.sort_values('cidade', inplace=True)
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x='pais', y='papel', data=principais_cidades_mundo_df, palette='crest')
plt.title('Principais Países - papel')
plt.xticks(rotation=90)
plt.show()
```
