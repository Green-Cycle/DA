```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from IPython.display import display, HTML
```

# Preparação dos dados

```python
porcentagem_por_regiao = r'Porcentagem_de_reciclagem_por_regiao.xlsx'

porcentagem_por_regiao = pd.read_excel(porcentagem_por_regiao)
```

```python
porcentagem_por_regiao.head()
```

```python
porcentagem_por_regiao.info()
```

```python
porcentagem_por_regiao.describe()
```

```python
porcentagem_por_regiao = porcentagem_por_regiao.rename(columns={"Região":"regiao", "Latas de Alumínio":"latas_alum", "Papel":"papel", 
                                                                "Plástico": "plastico", "Vidro": "vidro", "Papelão Ondulado": "papelao_ond", 
                                                                "Embalagens Longa Vida": "emb_lon_vd"})
```

```python
porcentagem_por_regiao.isna().sum()
```

```python
porcentagem_por_regiao.duplicated().sum()
```

# EDA

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))
sns.barplot(x='regiao', y='papel', data=porcentagem_por_regiao, palette='crest')
plt.title('Principais Regiões - papel')
plt.xticks(rotation=45)
plt.show()
```

A região que tem a maior porcentagem de reciclarem de papel é a Sudestee a menor é a Norte.

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))
sns.barplot(x='regiao', y='latas_alum', data=porcentagem_por_regiao, palette='crest')
plt.title('Principais Regiões - latas de alumínio')
plt.xticks(rotation=45)
plt.show()
```

As regiões reciclam latas de alumínio igualmente.

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 8))
sns.barplot(x='regiao', y='plastico', data=porcentagem_por_regiao, palette='crest')
plt.title('Principais Regiões - plástico')
plt.xticks(rotation=45)
plt.show()
```

A região que tem a maior porcentagem de reciclarem de plástico é a Sudeste e a menor é a Norte.

A região que tem a maior porcentagem de reciclarem de vidro é a Sudeste e a menor é a Norte.

A região que tem a maior porcentagem de reciclarem de papelão ondulado é a Sudeste e a menor é a Norte.
